```python
# A library of common molecules and their equilibrium geometries (in Angstroms)
molecule_library = {
    "hydrogen": "H 0.0 0.0 0.0; H 0.0 0.0 0.735",
    
    "ozone": (
        "O 0.0000  0.0000  0.0000; "
        "O 0.0000  1.0775  0.6865; "
        "O 0.0000 -1.0775  0.6865"
    ),
    
    "water": (
        "O 0.0000 0.0000 0.0000; "
        "H 0.0000 0.7572 -0.5865; "
        "H 0.0000 -0.7572 -0.5865"
    ),

    "carbon-dioxide": "C 0.0 0.0 0.0; O -1.16 0.0 0.0; O 1.16 0.0 0.0"
}
```

```python
import time
import numpy as np
from IPython.display import clear_output
from qiskit_nature.second_q.drivers import PySCFDriver
from qiskit_nature.second_q.mappers import JordanWignerMapper
from qiskit_algorithms import NumPyEigensolver
from qiskit_nature.second_q.algorithms import ExcitedStatesEigensolver
from qiskit_nature.second_q.transformers import ActiveSpaceTransformer

# 2. Pick your target molecule from the library
target_molecule = "water"
selected_geometry = molecule_library[target_molecule]

# 3. Initialize the driver using the library geometry selection
driver = PySCFDriver(atom=selected_geometry, basis="sto-3g")
raw_problem = driver.run()

transformer = ActiveSpaceTransformer(num_electrons=4, num_spatial_orbitals=4)
problem = transformer.transform(raw_problem)

# 4. Clear the screen of PySCF generation noise consistently
clear_output(wait=True)
time.sleep(0.1)

# 5. Set up your solvers and map the problem
mapper = JordanWignerMapper()
algo = NumPyEigensolver(k=100)
algo.filter_criterion = problem.get_default_filter_criterion()

solver = ExcitedStatesEigensolver(mapper, algo)
result = solver.solve(problem)

# 6. Dynamically print results based on your selection
print(f"--- Energy Spectrum for {target_molecule.upper()} ---")
print(f"Total ground state energy          = {result.total_energies[0]:.4f} Hartree")
print(f"Total first excited state energy   = {result.total_energies[1]:.4f} Hartree")
print(f"Total second excited state energy  = {result.total_energies[2]:.4f} Hartree")
print(f"Total third excited state energy   = {result.total_energies[3]:.4f} Hartree")
print(f"Total fourth excited state energy  = {result.total_energies[4]:.4f} Hartree")
```
