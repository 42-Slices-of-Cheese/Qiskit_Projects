```python
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister, transpile
from qiskit.visualization import plot_histogram
from qiskit_aer import Aer

# 3 data qubits + 2 ancilla for syndrome measurements
data = QuantumRegister(3, name='data')
anc = QuantumRegister(2, name='anc')
c_anc = ClassicalRegister(2, name='c_anc')  # for syndrome bits
c_out = ClassicalRegister(1, name='c_out')  # for final output
qc = QuantumCircuit(data, anc, c_anc, c_out)

# ----- Encode logical |+> state (superposition) -----
qc.h(data[0])
qc.cx(data[0], data[1])
qc.cx(data[0], data[2])

# ----- (Optional) Introduce a bit-flip error on data qubit 1 -----
qc.x(data[1])  # simulate an X error

# ----- Syndrome extraction -----
# Compare data[0] and data[1] using anc[0]
qc.cx(data[0], anc[0])
qc.cx(data[1], anc[0])
qc.measure(anc[0], c_anc[0])  # syndrome 1

# Compare data[1] and data[2] using anc[1]
qc.cx(data[1], anc[1])
qc.cx(data[2], anc[1])
qc.measure(anc[1], c_anc[1])  # syndrome 2

# ----- Final majority vote readout -----
qc.barrier()
qc.measure(data[1], c_out[0])  # pick one data qubit for final logical output

# ----- Run on simulator -----
sim = Aer
qc = transpile(qc, sim)
result = sim.execute(qc, sim, shots=1024).result()
counts = result.get_counts()

# Show result
plot_histogram(counts)
```
